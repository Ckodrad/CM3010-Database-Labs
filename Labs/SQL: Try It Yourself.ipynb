```python

%%sql
# Writing a query to list titles of books that were borrowed in the year 2022
SELECT b.Title, br.BorrowDate
FROM Books b
INNER JOIN Borrow br ON b.BookID = br.BookID
WHERE YEAR(br.BorrowDate) = 2022
```

```python
%%sql
# Writing query to display the names of members alongside the titles of books they have borrowed

SELECT b.Title, m.Name
FROM Books b
INNER JOIN Borrow br ON b.BookID = br.BookID
INNER JOIN Members m ON br.MemberID = m.MemberID;
```
